### Librerie utilizzate 

alcuni grafici non disponibili in Seaborn son stati visualizzati con matplotlib

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_excel('olive_oil.xlsx') #caricamento dati. NOTA: i nomi son stati ripuliti a mano dagli spazi in più presenti nel file Excel
df.columns #Aggiunto il nome "Name" alla prima colonna che è la colonna che contiene i nomi che classificano i campioni in 5 tipologie, NA, SA, U ,EL ,WL
```

```python
plt.figure(figsize=(10, 6))
pd.plotting.parallel_coordinates(df,'Name', color=('#2776f5','#f52f1d','#fafa19','#11bd08','#21fbff'))
plt.show()
```

***COMMENTO*** dal grafico delle Coordinate parallele non normalizzato non si notano particolari parametri in grado di dividere completamente i 5 tipi analizzati.

Si può comunque dire che la maggior divisione la si può notare nel parametro *Oleico* , *Palmitico* e *Linoleico*

```python
sns.pairplot(df, hue="Name",diag_kind="hist")
```

***COMMENTO*** 
Da questa plot matrix, in particolare dagli istogrammi, si nota come i dati siano spesso sovrapposti, ma si possa distinguere chiaramente 
il tipo SA (arancione) con un andamento simil gaussiano.

Infine, il parametro che ci dà una maggior divisione tra i gruppi e *SA* è l'*Oleico*.  Invece *Palmitico* ci permette dall'istogramma di osservare già 4 gruppi

```python
sns.boxplot(data=df)
```

***COMMENTO*** Il Boxplot mostra chiaramente come i dati siano perlopiù sovrapposti

```python
#df.boxplot(column=['oleico'], by='Name')
fig, axes = plt.subplots(4, 2, figsize=(20,18))
sns.boxplot(data=df, x='Name', y='palmitico', ax=axes[0,0])
sns.boxplot(data=df, x='Name', y='palmitoleico',ax=axes[0,1])
sns.boxplot(data=df, x='Name', y='stearico',ax=axes[1,0])
sns.boxplot(data=df, x='Name', y='oleico',ax=axes[1,1])
sns.boxplot(data=df, x='Name', y='linoleico',ax=axes[2,0])
sns.boxplot(data=df, x='Name', y='eicosanoico',ax=axes[2,1])
sns.boxplot(data=df, x='Name', y='linolenico',ax=axes[3,0])
```

***COMMENTO*** : Si può notare dal plotbox  che l'unico che divide i 5 tipi di olio è quello del parametro *Oleico* dove vi è solamente il campione SA diverso dagli altri

```python
sns.jointplot(data=df, x="oleico", y="palmitico",hue="Name")
```

```python
import math
nRows = len(df)
binDim = math.ceil(math.sqrt(nRows))
print(binDim)
```

```python
fig, axes = plt.subplots(4, 2, figsize=(20,15))
sns.histplot(df,x='palmitico',bins=binDim,ax=axes[0,0])
sns.histplot(df,x='palmitico',bins=2*binDim,ax=axes[0,0])

sns.histplot(df,x='palmitoleico',bins=binDim, ax=axes[0,1])
sns.histplot(df,x='palmitoleico',bins=2*binDim, ax=axes[0,1])


sns.histplot(df,x='stearico',bins=binDim, ax=axes[1,0])
sns.histplot(df,x='stearico',bins=2*binDim, ax=axes[1,0])

sns.histplot(df,x='oleico',bins=binDim, ax=axes[1,1])
sns.histplot(df,x='oleico',bins=2*binDim, ax=axes[1,1])

sns.histplot(df,x='linoleico',bins=binDim, ax=axes[2,0])
sns.histplot(df,x='linoleico',bins=2*binDim, ax=axes[2,0])

sns.histplot(df,x='eicosanoico',bins=binDim, ax=axes[2,1])
sns.histplot(df,x='eicosanoico',bins=2*binDim, ax=axes[2,1])

sns.histplot(df,x='linolenico',bins=binDim, ax=axes[3,0])
sns.histplot(df,x='linolenico',bins=2*binDim, ax=axes[3,0])
```

```python
# Compute the correlation matrix
corr = df.corr()
sns.color_palette("Spectral", as_cmap=True)
f, ax = plt.subplots(figsize=(20,10))
sns.heatmap(corr, cmap="jet")
```

Si può notare una correlazione maggiore tra palmitoleico-plmitico ed eicosanoico-linolenico
